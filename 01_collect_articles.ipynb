```python
import numpy as np
import pandas as pd
import pickle

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait

import re
import string
import time
from datetime import datetime,timedelta

import yfinance as yf
```

```python
print("Enter your ticker symbol:")
ticker = input()
```

```python
stock = yf.Ticker(ticker).history(start='2020-07-01', end=datetime.today())
```

```python
url = f'https://seekingalpha.com/symbol/{ticker}/news'
driver = webdriver.Chrome('./chromedriver')
driver.get(url)
```

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
links = driver.find_elements_by_tag_name('article')
```

```python
link_list = [i.find_element_by_tag_name('a').get_attribute('href') for i in links]
```

```python
df_list = []

for hyper_link in link_list:
    df_dict = {}
    time.sleep(5)
    driver.get(hyper_link)
    df_dict['date'] = driver.find_element_by_tag_name('time').text
    df_dict['title'] = driver.find_element_by_tag_name('h1').text
    mystr = ''
    for i in driver.find_elements_by_tag_name('p'):
        mystr = mystr + ' ' + i.text
    df_dict['text'] = mystr
    df_list.append(df_dict)
```

```python
df = pd.DataFrame(df_list)
```

```python
df = df[df['text'] != '']
```

```python
driver.close()
```

```python
df['title'] = [re.sub('[^A-Za-z0-9 $]+', '', i) for i in df['title']]
```

```python
df['text'] = [re.sub('[^A-Za-z0-9 $]+', '', i) for i in df['text']]
```

```python
df['new_date'] = [i[0:13] for i in df['date']]
df['new_date'] = pd.to_datetime(df['new_date'])
```

```python
df.set_index(df['new_date']).drop(columns='new_date').sort_index().join(stock,how='inner')
```
